```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_torus(precision, c, a): # num points, size, thickness
    #variables to loop through
    U = np.linspace(0, 2*np.pi, precision)
    V = np.linspace(0, 2*np.pi, precision)
    u, v = np.meshgrid(U, V)
    
    #parametric eqs
    x = (c + a * np.cos(v)) * np.cos(u)
    y = (c + a * np.cos(v)) * np.sin(u)
    z = a * np.sin(v)

    return x, y, z
```

```python
x, y, z = make_torus(10, 2, 1)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(z, x, y, color='yellow')

fig2 = plt.figure()
ax2 = fig2.add_subplot(projection='3d')
ax2.plot_wireframe(z, x, y)
```

```python
#take a look at each set of points (each ring)
fig = plt.figure()
ax2 = fig.add_subplot(projection='3d')
ax2.plot_wireframe(x, y, z, color='black',alpha=0.2)

#going thru each ring
for i in range(10):
    ax2.scatter(x[i], y[i], z[i])

#plotting bs dw ab it
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.view_init(elev=20., azim=-35, roll=0)
plt.show()
```

```python
import matplotlib as mpl
from matplotlib import _api, cbook, _docstring, _preprocess_data
def plot_surface(X, Y, Z, *, norm=None, vmin=None,
                     vmax=None, lightsource=None,):
        if Z.ndim != 2:
            raise ValueError("Argument Z must be 2-dimensional.")

        Z = cbook._to_unmasked_float_array(Z)
        X, Y, Z = np.broadcast_arrays(X, Y, Z)
        rows, cols = Z.shape

    
        rstride = 1
        cstride = 1

        colset = []  # the sampled facecolor
        if (rows - 1) % rstride == 0 and \
           (cols - 1) % cstride == 0 and \
           fcolors is None:
            polys = np.stack(
                [cbook._array_patch_perimeters(a, rstride, cstride)
                 for a in (X, Y, Z)],
                axis=-1)
        else:
            # evenly spaced, and including both endpoints
            row_inds = list(range(0, rows-1, rstride)) + [rows-1]
            col_inds = list(range(0, cols-1, cstride)) + [cols-1]

            polys = []
            for rs, rs_next in zip(row_inds[:-1], row_inds[1:]):
                for cs, cs_next in zip(col_inds[:-1], col_inds[1:]):
                    ps = [
                        # +1 ensures we share edges between polygons
                        cbook._array_perimeter(a[rs:rs_next+1, cs:cs_next+1])
                        for a in (X, Y, Z)
                    ]
                    # ps = np.stack(ps, axis=-1)
                    ps = np.array(ps).T
                    polys.append(ps)


        # In cases where there are non-finite values in the data (possibly NaNs from
        # masked arrays), artifacts can be introduced. Here check whether such values
        # are present and remove them.
        if not isinstance(polys, np.ndarray) or not np.isfinite(polys).all():
            new_polys = []
            new_colset = []

            # Depending on fcolors, colset is either an empty list or has as
            # many elements as polys. In the former case new_colset results in
            # a list with None entries, that is discarded later.
            for p, col in itertools.zip_longest(polys, colset):
                new_poly = np.array(p)[np.isfinite(p).all(axis=1)]
                if len(new_poly):
                    new_polys.append(new_poly)
                    new_colset.append(col)

            # Replace previous polys and, if fcolors is not None, colset
            polys = new_polys

        return polys

```

```python
import matplotlib.tri as mtri
from scipy.spatial import Delaunay
from scipy.spatial import ConvexHull, convex_hull_plot_2d
```

```python
def make_torus_tri(precision, c, a): # num points, size, thickness
    #variables to loop through
    U = np.linspace(0, 2*np.pi, precision)
    V = np.linspace(0, 2*np.pi, precision)
    u, v = np.meshgrid(U, V)
    
    u, v = u.flatten(), v.flatten()
    
    #parametric eqs
    x = (c + a * np.cos(v)) * np.cos(u)
    y = (c + a * np.cos(v)) * np.sin(u)
    z = a * np.sin(v)

    return x, y, z, u, v
```

```python
# LETS MAKE SOME TRIANGLES BABYYYY
tx, ty, tz, u, v = make_torus_tri(10, 2, 1)
tri = mtri.Triangulation(u, v)

fig = plt.figure()
ax2 = fig.add_subplot(projection='3d')
ax2.plot_trisurf(tx, ty, tz, triangles=tri.triangles, cmap=plt.cm.Spectral)

    
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.view_init(elev=30., azim=-35, roll=0)
plt.show()
```
